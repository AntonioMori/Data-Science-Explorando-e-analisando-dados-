```python
contagem_de_linguas = tmdb["original_language"].value_counts().to_frame().reset_index() #to frame para transformar em um novo framedata
#reset index, quando passa para um frame ele tenta transcrever de maneira automática o que erra alguns nomes das colunas, nesse caso colocou as linguas como index, o que
#não é totalmente errado
contagem_de_linguas.columns = ["original_language", "Total"]
#setando colunas
#primeiro dataframe criado      
```

```python
contagem_de_linguas.head()
```

```python
# Plotagem passando os parâmetros das propriedades de legenda
sns.barplot(data = contagem_de_linguas, x="original_language", y= "Total")
plt.title("")
plt.show()
```

```python
sns.countplot(data=tmdb, x="original_language")
#mesmo jeito porém utilizando já a função do seaborn chamada CountPlot
```

```python
#Como plotar um gráfico tipo pizza
contagem_de_linguas.plot(kind="pie",y= "Total",labels=contagem_de_linguas["original_language"])
```

```python
total_por_lingua = tmdb["original_language"].value_counts()
total_geral = total_por_lingua.sum()
total_de_ingles = total_por_lingua.loc["en"]
total_do_resto = total_geral - total_de_ingles

print(total_geral,total_de_ingles,total_do_resto)

```

```python
dados = {
  "lingua" : ["ingles", "outros"],
  "total" : [total_de_ingles, total_do_resto]
}
dados = pd.DataFrame(dados)
dados
```

```python
sns.barplot(x="lingua", y="total", data = dados)
```
