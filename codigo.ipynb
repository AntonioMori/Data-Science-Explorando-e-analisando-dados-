```python
contagem_de_linguas = tmdb["original_language"].value_counts().to_frame().reset_index() #to frame para transformar em um novo framedata
#reset index, quando passa para um frame ele tenta transcrever de maneira automática o que erra alguns nomes das colunas, nesse caso colocou as linguas como index, o que
#não é totalmente errado
contagem_de_linguas.columns = ["original_language", "Total"]
#setando colunas
#primeiro dataframe criado      
```

```python
contagem_de_linguas.head()
```

```python
# Plotagem passando os parâmetros das propriedades de legenda
sns.barplot(data = contagem_de_linguas, x="original_language", y= "Total")
plt.title("")
plt.show()
```

```python
total_por_lingua = tmdb["original_language"].value_counts()
total_geral = total_por_lingua.sum()
total_de_ingles = total_por_lingua.loc["en"]
total_do_resto = total_geral - total_de_ingles

print(total_geral,total_de_ingles,total_do_resto)

```

```python
dados = {
  "lingua" : ["ingles", "outros"],
  "total" : [total_de_ingles, total_do_resto]
}
dados = pd.DataFrame(dados)
dados
```

```python
sns.barplot(x="lingua", y="total", data = dados)
```

```python
total_de_outros_filmes_por_lingua = tmdb.query("original_language != 'en'")["original_language"].value_counts()
```

```python
plt.figure(figsize=(16,8))
sns.countplot(
                data=tmdb.query("original_language != 'en'"),
                x="original_language",
                order = total_de_outros_filmes_por_lingua.index, #ordenação
                hue="original_language", #agrupamento de cores de acordo com a coluna
                hue_order=total_de_outros_filmes_por_lingua.index, #ordenação das cores
                palette="mako", #paleta de cores
                stat="percent" #estatística
              )
plt.title("Quantidade de filmes exceção do inglês por língua")
plt.show()
```

```python

# Filtrando os dados para as colunas desejadas
dados = tmdb[["title", "original_language", "budget", "revenue"]]
linguas = ["fr", "es", "de"]

# Mantendo dados com valores de receita e orçamento válidos e nas línguas desejadas (fr, es, de)
fr_es_de_com_revenue_e_budget = dados.query("revenue > 0 and budget > 0 and original_language == @linguas")

# Construindo o gráfico de dispersão mais a reta de regressão linear para cada caso
sns.lmplot(data=fr_es_de_com_revenue_e_budget, x="budget", y="revenue", col="original_language", 
           hue="original_language", ci=None, height=4, scatter_kws={"s": 30, "alpha": 0.5})
```
